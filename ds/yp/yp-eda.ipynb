```python
def get_nullstat(df, columns): # —Ö–æ—Ç—å –¥–∞–Ω–Ω—ã–µ –≤–µ—Ä–Ω–µ—Ç, –≤ –æ—Ç–ª–∏—á–∏–µ –æ—Ç dataframe.info()
    df_result = pd.DataFrame(index=columns)
    for column in columns:
        df_stat = df[column].isnull().value_counts().to_frame(column)
        for value in [True, False]: # –µ—Å–ª–∏ –Ω—É–ª–µ–π –Ω–µ—Ç –∏–ª–∏ –≤—Å–µ –Ω—É–ª–∏, –∑–Ω–∞—á–µ–Ω–∏–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –æ–¥–Ω–æ, –∏—Å–ø—Ä–∞–≤–ª—è–µ–º
            if value not in df_stat.index:
                app = pd.DataFrame(index=[value],data=[0],columns=[column])
                df_stat = df_stat.append(app)
        df_line = pd.DataFrame(index=[column])
        for value in [True, False]:
            vf = df_stat[df_stat.index == value]
            vp = vf.pivot_table(index=vf.columns,columns=vf.index)
            vp.columns = [value]
            df_line[value] = vp[value]
            df_line["{} %".format(value)] = vp[value] / df_stat[column].sum() * 100
        df_result = df_result.append(df_line).dropna()
    return df_result
```

```python
def get_nullstat_agg(df, column_name, agg_column_name):
    result_null = df[df[column_name].isnull()][agg_column_name].value_counts().to_frame().reset_index()
    result_notnull = df[df[column_name].notnull()][agg_column_name].value_counts().to_frame().reset_index()
    result_detail = result_notnull.merge(result_null, on='index', how='left')
    result_detail.columns = [agg_column_name,"count({}) != null".format(column_name), "count({}) == null".format(column_name)]
    return result_detail
```

```python
nullstat = get_nullstat(data, data.columns)
nullstat[nullstat[True] == 0]
```

```python
def show_plot(df, xlab=None, ylab=None, name=None, ptype=['hist', 10], range=None, figsize=None):
    kvargs = {
        "title": name,
        "kind": ptype,
        "grid": True
    }
    if not figsize is None:
        kvargs['figsize'] = figsize
    if not range is None:
        kvargs['range'] = range
    if ptype[0] == 'hist':
        kvargs['kind'] = 'hist'
        kvargs['bins'] = ptype[1]
    if ptype[0] == 'line':
        kvargs['kind'] = 'line'
    if ptype[0] == 'box':
        kvargs['kind'] = 'box'
        kvargs['vert'] = ptype[1]
    if ptype[0] == 'scatter':
        kvargs['kind'] = 'scatter'
        kvargs['alpha'] = ptype[1]
        kvargs['x'] = ptype[2]
        kvargs['y'] = ptype[3]
        

    ax = df.plot(**kvargs)
    ax.set_xlabel(xlab)
    ax.set_ylabel(ylab)    
```

–ù–æ—Ä–º–∞–ª–∏–∑—É–µ–º **locality_name**

```python
unique_localities = set() # —Å–æ–±–µ—Ä–µ–º —Å—é–¥–∞ —Ç–æ–ª—å–∫–æ "–ø–æ—Å–ª–µ–¥–Ω–∏–µ" —Ç–æ–∫–µ–Ω—ã 
unique_loctypes = set()
final_loctypes = {}
def tokenize(x, update=False):
    if pd.isnull(x['locality_name']):
        if update:
            x['locality_type'] = x['locality_name'] # null - –ø–æ–∫–∞ –Ω–µ —Ç—Ä–æ–≥–∞–µ–º, –º–æ–∂–µ—Ç –ø–æ –≥–µ–æ–¥–∞–Ω–Ω—ã–º —É–¥–∞—Å—Ç—Å—è –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–∏—Ç—å
        return x
    cleared_lemmas = []
    value = x['locality_name'].lower().strip()
    # –Ω–∞ –≤—Ç–æ—Ä–æ–º –ø—Ä–æ–≥–æ–Ω–µ –º—ã —Å–æ–±–µ—Ä–µ–º —ç—Ç–æ—Ç —Å–ª–æ–≤–∞—Ä—å, –ø–æ–∫–∞ –æ–Ω –ø—É—Å—Ç–æ–π
    if update:
        for final_loctype in final_loctypes:
            if value.startswith(final_loctype):
                x['locality_type'] = final_loctypes[final_loctype].strip()
                x['locality_name'] = value[len(final_loctype):].strip()
                return x # —Ü–µ–ª–µ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
        # –µ—Å–ª–∏ –º—ã —Ç—É—Ç - –∑–Ω–∞—á–∏—Ç –Ω–∏—á–µ–≥–æ –Ω–µ –Ω–∞—à–ª–∏ –∏ –Ω–∞–¥–æ –≤–µ—Ä–Ω—É—Ç—å —Ç–µ–∫—É—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∞ –≤ locality_type –ø–µ—Ä–µ–¥–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é
        x['locality_type'] = '–≥–æ—Ä–æ–¥'
        x['locality_name'] = value.strip()
        return x
    lemmas = value.split(' ')
    for i in lemmas:
        lemma = i.strip() # –Ωe –Ω—É–∂–Ω—ã –ø—É—Å—Ç—ã–µ
        if(len(lemma) > 0):
            cleared_lemmas.append(lemma)
    unique_localities.add(cleared_lemmas.pop())
    loc = ' '.join(cleared_lemmas).strip()
    if len(loc) > 0: 
        unique_loctypes.add(loc)
    return x
data[['locality_name']].apply(lambda x: tokenize(x), axis=1)
print( unique_loctypes)
```

```python
def resurrect_by_nearest(x):
    if pd.isna(x['airports_nearest']) and pd.isna(x['cityCenters_nearest']): # –≤—Å—è —Å—Ç—Ä–æ–∫–∞ = nan
        return x
    airports_nearest = x['airports_nearest']
    cityCenters_nearest = x['cityCenters_nearest']   
    resurrectable_merge_row = resurrectable_merge.query('(airports_nearest == @airports_nearest) and (cityCenters_nearest == @cityCenters_nearest)')
    
    for column in ['locality_name','locality_type','parks_around3000','parks_nearest','ponds_around3000','ponds_nearest']:
        if pd.isna(x[column]):
            x[column] = resurrectable_merge_row[column + '_y'].values[0] # –Ω–µ –≤–∞–∂–Ω–æ —á—Ç–æ —Ç–∞–º, –¥–∞–∂–µ –µ—Å–ª–∏ null
    return x
resurrected_localities = resurrectable_localities.apply(resurrect_by_nearest, axis=1)
resurrected_localities
```

```python
data['floor_group'] = '–¥—Ä—É–≥–æ–π' # –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é
def floor_grouping(x):
    if pd.notnull(x['floor']):
        if x['floor'] == 1:
            x['floor_group'] = '–ø–µ—Ä–≤—ã–π'
        if pd.notnull(x['floors_total']):
            if x['floor'] == x['floors_total']:
                x['floor_group'] = '–ø–æ—Å–ª–µ–¥–Ω–∏–π'
    return x
data = data.apply(floor_grouping, axis=1)
data[['floor_group','total_area']].pivot_table(index='floor_group',values='total_area',aggfunc='count')
```
